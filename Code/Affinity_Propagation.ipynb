### üá¨üáß

This code block uses the `make_blobs` function to create an artificial dataset for clustering.
- `n_samples=100`: Generates 100 data points.
- `cluster_std=1`: Sets the standard deviation of the clusters, controlling how spread out they are.
- `random_state=0`: Ensures that the generated data is the same each time the code is run.
The function returns the generated data (`df4`) and the true labels for each point (`labels_true`).

```python
"""
Este bloco de c√≥digo aplica o algoritmo Affinity Propagation aos dados gerados.
- `preference=-30`: Define o par√¢metro de prefer√™ncia, que influencia o n√∫mero de clusters encontrados. Valores menores levam a menos clusters.
- `.fit(df4)`: Executa o algoritmo de clustering nos dados armazenados em `df4`.

As vari√°veis `cluster_centers_indices` e `labels` s√£o armazenadas para uso posterior.

This code block applies the Affinity Propagation algorithm to the generated data.
- `preference=-30`: Sets the preference parameter, which influences the number of clusters found. Lower values lead to fewer clusters.
- `.fit(df4)`: Executes the clustering algorithm on the data stored in `df4`.

The variables `cluster_centers_indices` and `labels` are stored for later use.
"""
af = AffinityPropagation (preference = -30, random_state = 0). fit(df4)
cluster_centers_indices = af.cluster_centers_indices_
labels = af.labels_
```

```python
"""
Este bloco de c√≥digo avalia o desempenho do clustering usando v√°rias m√©tricas do scikit-learn.
- `n_clusters_`: Calcula o n√∫mero estimado de clusters encontrados.
- `mt.homogeneity_score`, `mt.completeness_score`, `mt.v_measure_score`, `mt.adjusted_rand_score`, `mt.adjusted_mutual_info_score`: Calculam diferentes m√©tricas que comparam os r√≥tulos do cluster encontrados com os r√≥tulos verdadeiros.
- `mt.silhouette_score`: Calcula o coeficiente de silhueta, que mede qu√£o bem cada ponto de dados se encaixa em seu pr√≥prio cluster em compara√ß√£o com outros clusters.

This code block evaluates the clustering performance using various scikit-learn metrics.
- `n_clusters_`: Calculates the estimated number of clusters found.
- `mt.homogeneity_score`, `mt.completeness_score`, `mt.v_measure_score`, `mt.adjusted_rand_score`, `mt.adjusted_mutual_info_score`: Calculate different metrics that compare the found cluster labels with the true labels.
- `mt.silhouette_score`: Calculates the silhouette coefficient, which measures how well each data point fits into its own cluster compared to other clusters.
"""
import sklearn.metrics as mt
n_clusters_ = len(cluster_centers_indices)
print ("N√∫mero estimado de clusters: %d" % n_clusters_)
print ("Homogeneidade: %0.3f" % mt.homogeneity_score(labels_true, labels))
print ("Completude: %0.3f" % mt.completeness_score(labels_true, labels))
print ("Medida-V: %0.3f" % mt.v_measure_score(labels_true, labels))
print("√çndice Rand Ajustado: %0.3f" % mt.adjusted_rand_score(labels_true, labels))
print ("Informa√ß√£o M√∫tua Ajustada: %0.3f" % mt.adjusted_mutual_info_score(labels_true, labels))
print ("Coeficiente de Silhueta: %0.3f" % mt.silhouette_score(df4, labels, metric="sqeuclidean"))
```

```python
"""
Este bloco de c√≥digo define um esquema de cores aproximado de "hexachrome" para uso na plotagem.
- `hexachrome_colors`: Uma lista de c√≥digos hexadecimais que representam cores que se assemelham a um esquema hexachrome.
- `plt.cycler(color=hexachrome_colors)`: Cria um ciclo de cores a partir da lista definida, que ser√° usado para colorir os diferentes clusters no gr√°fico.

This code block defines an approximate "hexachrome" color scheme for use in plotting.
- `hexachrome_colors`: A list of hexadecimal codes representing colors that resemble a hexachrome scheme.
- `plt.cycler(color=hexachrome_colors)`: Creates a color cycle from the defined list, which will be used to color the different clusters in the plot.
"""
# Approximate hexachrome colors
hexachrome_colors = [
    '#000000',  # Black
    '#FFFFFF',  # White
    '#FF0000',  # Red
    '#00FF00',  # Green
    '#0000FF',  # Blue
    '#FFFF00',  # Yellow
]
colors = plt.cycler(color=hexachrome_colors)
```

```python
"""
Este bloco de c√≥digo gera o gr√°fico de dispers√£o dos clusters e seus centros, conectando cada ponto de dados ao centro do seu cluster.
- O loop itera sobre cada cluster (`k`) e sua cor correspondente (`col`).
- `class_members`: Seleciona os pontos de dados que pertencem ao cluster atual.
- `cluster_center`: Obt√©m as coordenadas do centro do cluster atual.
- `plt.scatter(df4[class_members, 0], df4[class_members, 1], ...)`: Plota os pontos de dados do cluster.
- `plt.scatter(cluster_center[0], cluster_center[1], ...)`: Plota o centro do cluster.
- O loop interno itera sobre cada ponto de dados no cluster e plota uma linha conectando-o ao centro do cluster.

This code block generates the scatter plot of the clusters and their centers, connecting each data point to its cluster center.
- The loop iterates through each cluster (`k`) and its corresponding color (`col`).
- `class_members`: Selects the data points that belong to the current cluster.
- `cluster_center`: Gets the coordinates of the current cluster center.
- `plt.scatter(df4[class_members, 0], df4[class_members, 1], ...)`: Plots the data points for the cluster.
- `plt.scatter(cluster_center[0], cluster_center[1], ...)`: Plots the cluster center.
- The inner loop iterates through each data point in the cluster and plots a line connecting it to the cluster center.
"""
for k, col in zip(range(n_clusters_), colors):
    class_members = labels == k
    cluster_center = df4[cluster_centers_indices[k]]
    plt.scatter(
        df4[class_members, 0], df4[class_members, 1], color=col["color"], marker="."
    )
    plt.scatter(
        cluster_center[0], cluster_center[1], s=14, color=col["color"], marker="o"
    )
    for x in df4[class_members]:
        plt.plot(
            [cluster_center[0], x[0]], [cluster_center[1], x[1]], color=col["color"]
        )
```
